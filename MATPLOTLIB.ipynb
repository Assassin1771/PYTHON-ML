# MATPLOTLIB

Matplotlib is a low level graph plotting library in python that serves as a visualization utility. <br>
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,1000);
#linspace function here will create 1000 ponits between 0 and 10

y=np.sin(x);

plt.plot(x,y)
```

```python
plt.scatter(x,y)
# a lot of points so scatter not shown
```

```python
plt.scatter(x[::10],y[::10])
```

```python
plt.scatter(x[::10],y[::10],color="red")
```

```python
plt.plot(x,y,color="blue")
plt.plot(x,np.cos(x),color="red")
```
